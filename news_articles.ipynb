```python
import pandas as pd
# Import data here
prices = pd.read_csv('data/upload_DJIA_table.csv', parse_dates=['Date'], index_col='Date')

prices.head()
```

```python
prices.loc['2015':]
```

```python
prices.sort_index().loc['2015':]
```

```python

# Set data frequency to business daily
prices = prices.asfreq('B')

# Create 'lagged' and 'shifted'
prices['lagged'] = prices['Close'].shift(-90)
prices['shifted'] = prices['Close'].shift(90)

# Plot the google price series
prices[['Close', 'lagged', 'shifted']].plot(subplots=True)
plt.show()


```
